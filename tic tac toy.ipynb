```python
X_O="X"
def cheacker(counter):
    global X_O
    if counter>3:
        messagebox.showinfo("player"+X_O, "congratulations you won")
def checkrow(r):
    counter=0
    global X_O
    for i in range(6):
        if butons[r][i]["text"]==X_O:
            counter+=1
        elif butons[r][i]["text"]!=X_O  :
            counter=0
        if counter>3:
            break
    cheacker(counter)
def checkColumn(c):
    global X_O
    counter=0
    for i in range(6):
        if butons[i][c]["text"]==X_O :
            counter+=1
        elif butons[i][c]["text"]!=X_O  :
            counter=0
        if counter>3:
            break
    cheacker(counter)
def checkLeftDigonal(ld):
    global X_O
    counter=0
    for i in range(6):
        for j in range(6):
            if butons[i][j]["text"]==X_O and butons[i][j].lDigonal==ld:
                counter+=1
            elif butons[i][j]["text"]!=X_O and butons[i][j].lDigonal==ld :
                counter=0
            if counter>3:
                break
    cheacker(counter)
def checkRightDigonal(rd):
    global X_O
    counter=0
    for i in range(6):
        for j in range(6):
            if butons[i][j]["text"]==X_O and butons[i][j].rDigonal==rd :
                counter+=1 
            elif butons[i][j]["text"]!=X_O and butons[i][j].rDigonal==rd:
                counter=0
            if counter>3:
                break
    cheacker(counter)
def checkAll(c,r,ld,rd):
    checkColumn(c)
    checkrow(r)
    checkLeftDigonal(ld)
    checkRightDigonal(rd)
    
from tkinter import *
from tkinter import messagebox
root7=Tk()

r, c= 6, 6;
counter=0
butons = [[0 for x in range(c)] for y in range(r)] 
statue=True
def ButtonEvent(buton):
    global X_O
    if X_O=="X"and buton["text"]=="":
        buton["text"]="X"
      
        checkAll(buton.column,buton.row,buton.lDigonal,buton.rDigonal)
        X_O="O"
    elif X_O=="O" and buton["text"]=="":
        buton["text"]="O"
     
        checkAll(buton.column,buton.row,buton.lDigonal,buton.rDigonal)
        X_O="X"
    

#initilizing array of buttons 6*6
for i in range(r):
    for j in range(c): 
        butons[i][j]=Button(root7,fg="red",bg="white",height =5, width = 8,
                            command=lambda i=i,j=j:ButtonEvent(butons[i][j]))
        butons[i][j].grid(row=i,column=j)
        butons[i][j].row,butons[i][j].column=i,j        
        counter+=1
# detrmining right digonal
co=1
for i in range(6):
    y=5-i
    for j in range(co):
        butons[j][y].rDigonal=i
        y+=1     
    co+=1    
co=5
for i in range(5):
    x=i+1
    for j in range(co):
        butons[x][j].rDigonal=i+6
        x+=1
    co-=1
# detrmining left digonal
co=1
for i in range(6):
    c=i
    for j in range(co):
        butons[j][c].lDigonal=i
        c-=1
    co+=1    
co=5
c=5
for i in range(5):
    x=i+1
    c=5
    for j in range(co):
        butons[x][c].lDigonal=i+6
        x=x+1
        c-=1
    co-=1
root7.mainloop() 
```
