```python
from tkinter import *
from tkinter import messagebox
color="red"
def cheacker(counter):
    global color
    if counter>3:
        messagebox.showinfo("player "+color, color +" wins congratulations ")
        answer=messagebox.askquestion("new game","do you want to play again")
        if answer=="yes":
            for i in range(6):
                for j in range(6): 
                    butons[i][j].color="black"
                    canvas.itemconfig(butons[i][j].id, fill="black")
       
            
def checRow(r):
    counter=0
    global color
    for i in range(6):
        if butons[r][i].color==color:
            counter+=1
        elif butons[r][i].color!=color:#sıra bozulmuşsa
            counter=0
        if counter>3:
            break
    cheacker(counter)
def checkColumn(c):
    global colors
    counter=0
    for i in range(6):
        if butons[i][c].color==color:
            counter+=1
        #sıra bozulmuşsa
        elif butons[i][c].color!=color:
            counter=0
        if counter>3:
            break
    cheacker(counter)
def checkLeftDigonal(ld):
    global colors
    counter=0
    for i in range(6):
        for j in range(6):
            if butons[i][j].color==color and butons[i][j].lDigonal==ld:
                counter+=1
            elif butons[i][j].color!=color and butons[i][j].lDigonal==ld :
                counter=0
            if counter>3:
                break
    cheacker(counter)
def checkRightDigonal(rd):
    global colors
    counter=0
    for i in range(6):
        for j in range(6):
            if butons[i][j].color==color and butons[i][j].rDigonal==rd :
                counter+=1 
            elif butons[i][j].color!=color and butons[i][j].rDigonal==rd:
                counter=0
            if counter>3:
                break
    cheacker(counter)
class circle:
    def __init__(self,canvas,x,y, r):
        self.colum=x
        self.row=y
        self.color="black"
        x=100+x*100
        y=130+y*100
        self.id = canvas.create_oval(x-r,y-r,x+r,y+r,fill="black")
        canvas.tag_bind(self.id,"<Button-1>", callback2)
        
        

def callback2(event):#buton eventi
    global canvas
    global color
    global butons
    idList = canvas.find_overlapping(event.x, event.y, event.x, event.y)#seçilen itemin idis tuple olarak
    selected=gettingSelected(butons,idList[0])#seçilen butonu döduruyor yoksa -1 döndürür
    if selected!=-1:
        for i in range(5,-1,-1):# seçlen butonun sütününe bakiyor ve bulduğu ilk boş daireyi sıraya döre rengini değiştiriyor
            if butons[i][selected.colum].color=="black":
                canvas.itemconfig(butons[i][selected.colum].id, fill=color)
                butons[i][selected.colum].color=color
                selected=butons[i][selected.colum]
                break
        checkColumn(selected.colum)
        checRow(selected.row)
        checkRightDigonal(selected.rDigonal)
        checkLeftDigonal(selected.lDigonal)
        if color=="red":
            color="yellow"
        else:
            color="red"
    


root = Tk()
canvas= Canvas(root, width=700, height=700,bg="blue")
r, c= 6, 6;
counter=0
butons = [[0 for x in range(c)] for y in range(r)] 
for i in range(r):
    for j in range(c): 
        butons[j][i]=circle(canvas,i,j,40) 
pressed=butons[0][0].id
def gettingSelected(butons,id):
    for i in range(6):
        for j in range(6): 
            if butons[i][j].id==id:
                return butons[i][j]
    return -1;
# detrmining right digonal
co=1
for i in range(6):
    y=5-i
    for j in range(co):
        butons[j][y].rDigonal=i
        y+=1     
    co+=1    
co=5
for i in range(5):
    x=i+1
    for j in range(co):
        butons[x][j].rDigonal=i+6
        x+=1
    co-=1
# detrmining left digonal
co=1
for i in range(6):
    c=i
    for j in range(co):
        butons[j][c].lDigonal=i
        c-=1
    co+=1    
co=5
c=5
for i in range(5):
    x=i+1
    c=5
    for j in range(co):
        butons[x][c].lDigonal=i+6
        x=x+1
        c-=1
    co-=1

canvas.pack()

root.mainloop()     

```
